```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('dataset_soil.csv')
data
```

```python
data.info()
```

```python
real_x = data.iloc[:,:5].values
#real_x
```

```python
real_y = data.iloc[:,5].values
#real_y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x, test_x, train_y, test_y =  train_test_split(real_x,real_y,test_size=0.3, random_state=99)
```

```python
train_x.shape, train_y.shape
```

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
```

```python
le.fit(train_x,train_y)
```

```python
y_pred = le.predict(test_x)
#y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(test_y,y_pred)
```

```python
le.coef_
```

```python
le.intercept_
```
