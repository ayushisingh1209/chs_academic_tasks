```r
st<-c(19.5, 24.7, 30.7, 29.8, 19.1, 25.6, 31.4, 27.9, 22.1, 25.5, 31.1,
30.4, 18.7, 19.7, 14.6, 29.5, 27.7, 30.2, 22.7, 25.2)

tc<-c(43.1, 49.8, 51.9, 54.3, 42.2, 53.9, 58.5, 52.1, 49.9, 53.5, 56.6,
56.7, 46.5, 44.2, 42.7, 54.4, 55.3, 58.6, 48.2, 51.0)

mc<-c(29.1, 28.2, 37.0, 31.1, 30.9, 23.7, 27.6, 30.6, 23.2, 24.8, 30.0,
28.3, 23.0, 28.6, 21.3, 30.1, 25.7, 24.6, 27.1, 27.5)

bf<-c(11.9, 22.8, 18.7, 20.1, 12.9, 21.7, 27.1, 25.4, 21.3,
19.3, 25.4, 27.2, 11.7, 17.8, 12.8, 23.9, 22.6, 25.4, 14.8,
21.1)
```

```r
par(mfrow=c(2, 2))
plot(st,type = "o")
plot(tc,type = "o")
plot(mc,type = "o")
plot(bf,type = "o")
```

```r
st_tc <- st*tc
st_mc <- st*mc
tc_mc <- tc*mc
st_tc_mc <- st*tc*mc

fat <- data.frame(st, tc, mc, st_tc, st_mc, tc_mc, st_tc_mc, bf)
head(fat)
```

```r
par(mfrow=c(2, 2))
plot(st_tc,type = "o")
plot(st_mc,type = "o")
plot(tc_mc,type = "o")
plot(st_tc_mc,type = "o")
```

```r
summary(fat)
```

```r
pairs(fat)
```

```r
cor(fat)
```

```r
install.packages("ggcorrplot")
library(ggcorrplot)
corr <- round(cor(fat), 1)

# Plot
ggcorrplot(corr,
           type = "lower",
           lab = TRUE,
           lab_size = 5,
           colors = c("tomato2", "white", "springgreen3"),
           title="Correlogram of Housing Dataset",
           ggtheme=theme_bw)
```

```r
# Create the centered independent variables:
st.c <- st - mean(st)
tc.c <- tc - mean(tc)
mc.c <- mc - mean(mc)

# Now create the centered interaction terms:
st_tc.c <- st.c*tc.c
st_mc.c <- st.c*mc.c
tc_mc.c <- tc.c*mc.c
st_tc_mc.c <- st.c*tc.c*mc.c
```

```r
# Create a new data frame with this new set of independent variables
fat.c <- data.frame(st.c, tc.c, mc.c, st_tc.c, st_mc.c, tc_mc.c, st_tc_mc.c, bf)
head(fat.c)
```

```r
cor(fat.c)
```

```r
library(ggcorrplot)
corr <- round(cor(fat.c), 1)

# Plot
ggcorrplot(corr,
           type = "lower",
           lab = TRUE,
           lab_size = 5,
           colors = c("tomato2", "white", "springgreen3"),
           title="Correlogram of Housing Dataset",
           ggtheme=theme_bw)
```

```r
# Calculate a simple linear regression for each variable (not the interactions).
regression1.out <- lm(bf ~ st.c)
regression2.out <- lm(bf ~ tc.c)
regression3.out <- lm(bf ~ mc.c)
```

```r
print(paste("Regression 1 Intercept : ",coef(regression1.out)[1]))
print(paste("Regression 1 Slope : ",coef(regression1.out)[2]))
print("----------------------------------------------------------")
print(paste("Regression 2 Intercept : ",coef(regression2.out)[1]))
print(paste("Regression 2 Slope : ",coef(regression2.out)[2]))
print("----------------------------------------------------------")
print(paste("Regression 3 Intercept : ",coef(regression3.out)[1]))
print(paste("Regression 3 Slope : ",coef(regression3.out)[2]))
```

```r
par(mfrow=c(1, 4))
plot(regression1.out)
plot(regression2.out)
plot(regression3.out)
```

```r
regression4.out <- lm(bf ~ st.c + tc.c + mc.c)
summary(regression4.out)
```

```r
regression5.out <- lm(bf ~ st.c + mc.c)
summary(regression4.out)
```

```r
regression6.out <- lm(bf ~ st.c + mc.c + st_mc.c)
summary(regression6.out)
```

```r
par(mfrow=c(1, 4))
plot(regression4.out)
plot(regression5.out)
plot(regression6.out)
```
